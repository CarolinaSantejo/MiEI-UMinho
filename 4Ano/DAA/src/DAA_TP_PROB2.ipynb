## Corrigir valores em falta (NaN's)

```
def remove_nans(data, column, verbose=False):
    temp = data

    if verbose:
        print(f'Valores NaN: {temp[column].isnull().sum()}')
        print(f'Valores Não NaN: {temp[column].count() - temp[column].isnull().sum()}')
        print(f'Percentagem de valores nan: {temp[column].isnull().sum() / temp[column].count()}')
  
    temp.dropna(axis=0, inplace=True, subset=[column])

    return temp
```

## Converter Dados Categóricos ('Gender', 'Company Type', 'WFH')

```
def cat_to_ord_mapping(data):
    temp = data
    gender_mapper = {'Male':0, 'Female':1}
    temp = data['Gender'].replace(gender_mapper)

    company_type_mapper = {'Service': 0, 'Product': 1}
    temp = temp['Company Type'].replace(company_type_mapper)

    wfh_mapper = {'No': 0, 'Yes': 1}
    temp = temp['WFH Setup Available'].replace(wfh_mapper)
    
    return temp

def one_hot_encode(data):
    temp = data
    one_hot_gender = pd.get_dummies(temp['Gender'], prefix='Gender')
    temp = temp.join(one_hot_gender)

    one_hot_wfh = pd.get_dummies(data['WFH Setup Available'], prefix="WFH Setup")
    temp = temp.join(one_hot_wfh)

    temp.drop(['Gender', 'WFH Setup Available'], axis=1, inplace=True)
    
    return temp
```

```
data = one_hot_encode(data)
data
```

## Automatização do processo de tratamento de dados

```
def apply_data_treatment(data, keep_date=False):
    new_data = data


    new_data.drop('Employee ID', axis=1, inplace=True)
    new_data.drop('Company Type', axis=1, inplace=True)

    if keep_date==False:
        new_data.drop('Date of Joining', axis=1, inplace=True)

    new_data = remove_nans(data, 'Mental Fatigue Score', verbose=True)
    new_data = remove_nans(data, 'Resource Allocation', verbose=True)
    new_data = remove_nans(data, 'Burn Rate', verbose=True)

    new_data = one_hot_encode(data)

    return new_data
   
    
```

# Redes Neuronais

```
import tensorflow as tf
from tensorflow.keras.models import Sequential
from keras.layers import Dense, Dropout, BatchNormalization
from keras.wrappers.scikit_learn import KerasRegressor
from sklearn.model_selection import GridSearchCV, KFold, train_test_split
from sklearn.preprocessing import MinMaxScaler
```

## Preparar dados para utilização

```
data = pd.read_csv('train.csv')
data = apply_data_treatment(data)
data
```

## Normalização dos dados de treino/teste

```
X = data.drop(['Burn Rate'], axis=1)
y = data['Burn Rate'].to_frame()

X
```

```
y
```

```
scaler_X = MinMaxScaler(feature_range=(0,1)).fit(X)
X_scaled = pd.DataFrame(scaler_X.transform(X[X.columns]), columns=X.columns)

scaler_y = MinMaxScaler(feature_range=(0,1)).fit(y)
y_scaled = pd.DataFrame(scaler_y.transform(y[y.columns]), columns=y.columns)

X_scaled
```

```
y_scaled
```

## Criação do modelo

```
def build_model(activation='relu', learning_rate=0.01):
    model = Sequential()
    model.add(Dense(16, input_dim=7, activation=activation))
    model.add(Dense(8, activation=activation))
    model.add(Dense(1, activation=activation))

    model.compile(
        loss='mae',
        optimizer=tf.optimizers.Adam(learning_rate),
        metrics=['mae', 'mse']
    )

    return model
```

## Treino do Modelo

```
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size=0.3, random_state=2021)
```

```
kfold = KFold(n_splits=5, shuffle=True, random_state=2021)
paramGrid = {
    'activation': ['relu', 'sigmoid'],
    'learning_rate': [0.01, 0.02, 0.03]
}
```

```
model = KerasRegressor(build_fn=build_model, epochs=100, batch_size=128)
grid_search = GridSearchCV(estimator=model, param_grid=paramGrid, cv=kfold, scoring='neg_mean_absolute_error', refit=True, verbose=1)
```

```
grid_search_result = grid_search.fit(X_train, y_train, validation_split=0.3, verbose=1)
```

```

means = grid_search.cv_results_['mean_test_score']
stds = grid_search.cv_results_['std_test_score']
params = grid_search.cv_results_['params']

for mean, std, param in zip(means, stds, params):
    print(f'{mean} : {std} : {param}')
```

## Análise dos Resultados e do Desempenho

```
best_mlp_model = grid_search.best_estimator_
```

```
def plot_learning_curve(history, metric='neg_mean_absolute_error'):
    plt.figure(figsize=(10,10))
    plt.title('Training Loss vs Validation Loss')
    plt.plot(history.epoch, history.history['loss'], label='train')
    plt.plot(history.epoch, history.history['val_loss'], label='validation')
    plt.ylabel('loss ' + metric)
    plt.xlabel('Epochs')
    plt.legend()

plot_learning_curve(best_mlp_model.model.history, metric='neg_mean_absolute_error')
```

```
predictions = best_mlp_model.predict(X_test)
predictions
```

```
plt.xlabel('Dados de teste (y_test)')
plt.ylabel('Previsões (predictions)')
plt.scatter(y_test, predictions)
```

```
sns.displot((y_test-predictions.reshape(5577, 1)), bins=32, kde=True)
```

```
print(f'MAE: {mean_absolute_error(y_test, predictions)}')
print(f'MSE: {mean_squared_error(y_test, predictions)}')
print(f'RMSE: {np.sqrt(mean_squared_error(y_test, predictions))}')
```
